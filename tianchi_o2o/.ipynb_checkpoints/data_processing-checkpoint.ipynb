```python
def calc_discount_rate(s):
    """
    获取优惠券的折扣率，包括满减类型和折扣类型的优惠券
    """
    s =str(s)
    s = s.split(':')
    if len(s)==1:
        return float(s[0])
    else:
        return 1.0-float(s[1])/float(s[0])

def get_discount_man(s):
    """
    如果是满减类型的优惠券，获取满减条件
    """
    s =str(s)
    s = s.split(':')
    if len(s)==1:
        return 'null'
    else:
        return int(s[0])
        
def get_discount_jian(s):
    """
    如果是满减类型的优惠券，获取满减额度
    """
    s =str(s)
    s = s.split(':')
    if len(s)==1:
        return 'null'
    else:
        return int(s[1])

def is_man_jian(s):
    """
    判断优惠券是满减类型（置1）还是折扣类型（置0），
    """
    s =str(s)
    s = s.split(':')
    if len(s)==1:
        return 0
    else:
        return 1
```

>为所有的数据集添加优惠券折扣率、优惠券领取日期相关的特征

><a name="7">消费者相关的特征提取</a>  
<font color=red>统计期间该消费者消费的商户数：</font>count_merchant. 
<font color=red>统计期间该消费者距离使用优惠券消费的那些商户的平均距离/最小距离/最大距离：</font>user_avg_distance <font color=red>/</font> user_min_distance <font color=red>/</font> user_max_distance 
<font color=red>统计期间该消费者优惠券使用次数/总共的消费次数/优惠券领取次数：</font>buy_use_coupon  <font color=red>/</font> buy_total  <font color=red>/</font> coupon_received  
<font color=red>统计期间该消费者优惠券使用率</font>buy_use_coupon/coupon_received. 
<font color=red>统计期间该消费者使用优惠券消费次数占总消费次数的比率</font>buy_use_coupon/buy_total
<font color=red>统计期间该消费者：</font>user_date_datereceived_gap
<font color=red>feature3:</font>训练集中20160315-20160630期间消费（含用券和不用券）或者领了券但未消费的数据集  
<font color=red>feature2:</font>训练集中20160201-20160514期间消费（含用券和不用券）或者领了券但未消费的数据集  
<font color=red>feature1:</font>训练集中20160101-20160413期间消费（含用券和不用券）或者领了券但未消费的数据集

```python
def get_user_date_datereceived_gap(s):
    s = s.split(':')
    return (date(int(s[0][0:4]),int(s[0][4:6]),int(s[0][6:8])) - date(int(s[1][0:4]),int(s[1][4:6]),int(s[1][6:8]))).days
```

```python
def get_label(s):
    s = s.split(':')
    if s[0]=='null':
        return 0
    elif (date(int(s[0][0:4]),int(s[0][4:6]),int(s[0][6:8]))-date(int(s[1][0:4]),int(s[1][4:6]),int(s[1][6:8]))).days<=15:
        return 1
    else:
        return -1
```
