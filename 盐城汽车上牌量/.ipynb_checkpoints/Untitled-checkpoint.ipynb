```python
import pandas as pd
import numpy as np
import pyecharts as pe
import warnings
warnings.filterwarnings('ignore')
```

```python
trainset = pd.read_csv('./data/data/train_set_revised1.csv')
testsetA = pd.read_csv('./data/data/test_A_set_revised1.csv')

trainset = trainset[trainset['date']>0]
testsetA = testsetA[testsetA['date']>0]
dataset1 = trainset[['date','day_of_week','calendar','year','month','day_of_year','gap']].drop_duplicates()
testsetA1 = testsetA[['date','day_of_week','calendar','year','month','day_of_year','gap']].drop_duplicates()

def get_month_period(calendar):
    day = int(calendar.split('/')[2])
    if day<=10:
        return 1
    elif day<=20:
        return 2
    else:
        return 3
dataset1['period_of_month']=dataset1['calendar'].apply(get_month_period)
testsetA1['period_of_month']=testsetA1['calendar'].apply(get_month_period)

dataset1['total_cnt'] = trainset[['date','cnt']].groupby('date').sum()['cnt'].values

trainset = dataset1[['day_of_week','month','gap','period_of_month','total_cnt']]
testsetA = testsetA1[['day_of_week','month','gap','period_of_month']]
```

```python
q_list = []
for i in range(1,7):
    q_list.append(int(trainset.loc[trainset['day_of_week']==i,'total_cnt'].quantile(q=0.25)))
    q_list.append(int(trainset.loc[trainset['day_of_week']==i,'total_cnt'].quantile(q=0.75)))
for i in range(6):
    trainset.loc[(trainset['day_of_week']==i+1)&(trainset['total_cnt']<q_list[i*2]),'total_cnt'] = q_list[i*2]
    trainset.loc[(trainset['day_of_week']==i+1)&(trainset['total_cnt']>q_list[i*2+1]),'total_cnt'] = q_list[i*2+1]
```
